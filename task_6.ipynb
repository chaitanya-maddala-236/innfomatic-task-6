```python
def smart_parking_system(capacity, vehicle_logs):
    """
    Function to manage parking lot based on entry and exit logs.
    """

    parked = 0
    peak_usage = 0

    # Process each log
    for log in vehicle_logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        # Track peak usage
        peak_usage = max(peak_usage, parked)

    # Status check
    if parked > capacity:
        status = "Alert: Parking Exceeded Capacity!"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak_usage)
    print("Parking Status:", status)
```

```python
def delivery_time_estimator(distance, traffic, weather):
    """
    Estimate delivery time based on distance, traffic, and weather.
    """

    # Base time: assume 5 minutes per km
    eta = distance * 5

    # Traffic adjustment
    if traffic.lower() == "high":
        eta += 15
    elif traffic.lower() == "medium":
        eta += 8

    # Weather adjustment
    if weather.lower() == "rainy":
        eta += 10
    elif weather.lower() == "storm":
        eta += 20


    print("Estimated Delivery Time:", eta, "minutes")
delivery_time_estimator(8, "High", "Rainy")
```

```python
def theatre_occupancy(total_seats, booked_seats):
    """
    Analyze theatre seat occupancy.
    """

    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats Available"

    print(f"Occupancy: {occupancy:.0f}%")
    print("Show Status:", status)
total_seats = 200
booked = [1] * 150

theatre_occupancy(total_seats, booked)
```

```python
def server_load_classifier(cpu_readings):
    """
    Classify server load based on average CPU usage.
    """

    avg_cpu = sum(cpu_readings) / len(cpu_readings)

    if avg_cpu < 50:
        status = "Normal"
    elif 50 <= avg_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print(f"Average CPU Load: {avg_cpu:.0f}%")
    print("Server Status:", status)
readings = [45, 60, 70, 85, 90]
server_load_classifier(readings)
```

```python
def classroom_resource_monitor(resource_usage):
    """
    Identify overused classroom resources.
    """

    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:   # threshold
            overused.append(resource)

    if overused:
        alert = "Yes"
    else:
        alert = "No"

    print("Overused Resources:", ", ".join(overused) if overused else "None")
    print("Energy Alert:", alert)
```

```python
usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(usage)
```

```python
def event_registration_controller(capacity, registrations):
    """
    Manage event registrations and waitlist.
    """

    if registrations <= capacity:
        confirmed = registrations
        waitlisted = 0
        status = "Open"
    else:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)
```

```python
event_registration_controller(100, 105)
```
